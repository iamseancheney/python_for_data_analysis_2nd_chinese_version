### 快捷键
- enter 和 esc 命令模式和编辑模式互换
- shift enter 运行当前代码块并跳转到下一个代码块
- ctrl enter 运行但不跳转
- 命令模式下
    - m：变成md格式   
    - y：变成代码格式
    - a: 在上方下创建新的代码块
    - b：在下方创建新的代码块
    - l：在同一代码块中的多行代码加上行号
    - D D（连续按两次字母 D） → 删除当前单元格
